# Data Cleaning

Android Ver column

```python
df_copy['Android Ver'].isnull().sum()
```

```python
df_copy['Android Ver'].unique()
```

```python
# Splting

df_copy['Android Ver']=df_copy['Android Ver'].str.replace('and up','')
df_copy['Android Ver']=df_copy['Android Ver'].str.replace('- 7.1.1','')
```

```python
df_copy['Android Ver'].unique()
```

```python
df_copy['Android Ver']=df_copy['Android Ver'].replace('Varies with device',np.nan)
df_copy['Android Ver']=df_copy['Android Ver'].str.replace('W','')
```

```python
df_copy['Android Ver'].unique()
```

```python
df_copy['Android Ver'].isnull().sum()
```

```python
df_copy['Android Ver']=df_copy['Android Ver'].str.extract(r'(\d+\.\d+)', expand=False)
```

```python
df_copy['Android Ver'].isnull().sum()
```

```python
df_copy['Android Ver']=df_copy['Android Ver'].astype(float)
```

```python
df_copy.info()
```

Successfully cleaned and transformed the Android Ver column into float type

```python
sns.boxplot(df_copy['Android Ver'])
```

Observing the boxplot, we can say the Android Ver column contains many outliers. To fill the missing values, we will use the median of the Android Ver column

```python
df_copy['Android Ver']=df_copy['Android Ver'].fillna(df_copy['Android Ver'].median())
```

```python
df_copy['Android Ver'].isnull().sum()
```

All the missing values of the Android Ver column have been filled.

Creating a new feature: minimum age (Min_age).
This feature converts the Content Rating categories into a numeric age threshold, making it easier to analyze and use in machine‑learning models.

```python
# Defining Minimum age according to unique values of the Content Rating feature
age_gap={
'Everyone':0,
'Teen':13,
'Everyone 10+':10,
'Mature 17+':17,
'Adults only 18+':18,
'Unrated':np.nan
}

df_copy['Min age']=df_copy['Content Rating'].map(age_gap)
```

```python
# Dropping the content Rating column as we no longer need it

df_copy=df_copy.drop('Content Rating', axis=1)
```

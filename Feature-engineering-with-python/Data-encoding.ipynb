Nominal/OHE-endoding

```python
import numpy as np 
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.DataFrame({'color':['red','yellow','green','red', 'yellow']})
df
```

```python
# Performing fit and transform
encoder=OneHotEncoder()
encoded=encoder.fit_transform(df[['color']]).toarray()
```

```python
encoded_df=pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
```

```python
encoded_df
```

```python
df2=pd.DataFrame({'color':['red','purple','yellow','red']})
```

```python
from sklearn.preprocessing import LabelEncoder

encoder=LabelEncoder()

encoder.fit_transform(df2[['color']])
```

```python
encoder.transform([['red']])
```

```python
encoder.transform([['purple']])
```

ORDINAL ENCODING

```python
# Creating dataframe
df3=pd.DataFrame({'color':['red','magenta','lavendar','red','beige']})

from sklearn.preprocessing import OrdinalEncoder
```

```python
encoder=OrdinalEncoder(categories=[['red','magenta','lavendar','beige']])
```

```python
encoder.fit_transform(df3[['color']])
```

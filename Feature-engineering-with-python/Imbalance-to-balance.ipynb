```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('Cerebral Stroke Prediction-Imbalanced Dataset.csv')
df
```

```python
df['stroke'].value_counts(),df['stroke'].value_counts(normalize=True)# this gives the proportion
```

UNBALANCED

```python
# Splitting the majority and minority
df_major=df[df['stroke']==0]
df_minior=df[df['stroke']==1]
```

```python
# Upsample the minority to match the majority
from sklearn.utils import resample

df_minority_up=resample(
    df_minior,
    replace=True,               #sample with replacement
    n_samples=len(df_major),    # matching majority count
    random_state=42
)
# Combining
df_up=pd.concat([df_major, df_minority_up]).reset_index(drop=True)
df_up
```

```python
df_up['stroke'].value_counts(), df_up['stroke'].value_counts()
```

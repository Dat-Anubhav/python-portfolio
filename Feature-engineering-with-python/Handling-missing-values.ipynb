```python
import seaborn as sns

df=sns.load_dataset('titanic')

print(df)
```

```python
df.isnull().sum()
```

Missing data columns :- age, embarked, deck, embark_town
. Deck column has the highest number of missing values

```python
sns.histplot(df['age'], kde=True)
```

since age data is skewed means contain outliers therefore i will use median to fill missing values

```python
df
```

```python
df.head(20)
```

```python
df['age'].median()
```

```python
# Filling age column missing values with median of the column
df['age']=df['age'].fillna(df['age'].median())
df.head(30)
```

```python
# Now embarked column
sns.histplot(df['embarked'], kde=True)
```

```python
# Filling missing values for embarked column 
df['embarked']=df['embarked'].fillna(df['embarked'].mode()[0])
```

```python
# Filling missing values for deck column 
sns.histplot(df['deck'], kde=True)
# df['deck']=df['deck'].fillna(df['embarked'])
```

since deck column is a categorical data, to fill the missing value we have to use mode imputation

```python
df['deck']=df['deck'].fillna(df['deck'].mode()[0])
df
```

df['deck'].mode() returns a Series, not a single value.

Example structure: 0 C (and possibly more modes). The index 0 is just the first row of that Series.â€‹

When you write df['deck'].mode()[0], you are taking only the first element of that Series, so the result is a scalar value like 'C'.

fillna needs a scalar (or something broadcastable) to replace all NaNs; once you pass 'C' instead of the whole Series, pandas can fill every missing deck with that one category.â€‹

So [0] simply selects â€œthe first most frequent categoryâ€ from the list of modes and makes the imputation work as intended.

```python
# filling missing values of a embark_town column
sns.histplot(df['embark_town'], kde=True)
```

```python
df['embark_town']=df['embark_town'].fillna(df['embark_town'].mode()[0])
df
```

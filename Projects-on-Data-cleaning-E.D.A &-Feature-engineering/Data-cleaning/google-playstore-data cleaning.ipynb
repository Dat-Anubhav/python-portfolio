Moving towards the 'Size' Column

```python
df_copy['Size'].isnull().sum()
```

```python
df_copy['Size'].unique()
```

```python
# Cleaning the Size feature by removing the '+', ',', 'M', 'k' and 'Varies with device'
df_copy['Size']=df_copy['Size'].str.replace(',','')
df_copy['Size']=df_copy['Size'].str.replace('+','')
df_copy['Size']=df_copy['Size'].replace('Varies with device',np.nan)
df_copy['Size']=df_copy['Size'].str.replace('M','000k')
df_copy['Size']=df_copy['Size'].str.replace('k','000')
```

I was struggling with this column because the Size feature contains only a few entries with '+' and ',' and some entries contain only a comma. These characters break numeric conversion, so identifying them was difficult. Even after finding them, the order of cleaning matters a lot; first the commas and plus signs must be removed so that the values become valid numeric strings, and only then should the 'M' (millions) and 'k' (thousands) units be handled.

```python
df_copy['Size'].unique()
```

```python
# Converting Size column values into a float type

df_copy['Size']=df_copy['Size'].astype(float)
```

```python
# Finding number of Missing values
df_copy['Size'].isnull().sum()
```

```python
sns.histplot(df_copy['Size'],kde=True)
```

```python
sns.boxplot(df_copy['Size'])
```

By observing the boxplot, we can say that the Size feature contains many strong outliers; therefore, we will use the median to fill the NaN (missing values).

```python
# Filling the NAN values

df_copy['Size']=df_copy['Size'].fillna(df_copy['Size'].median())
```

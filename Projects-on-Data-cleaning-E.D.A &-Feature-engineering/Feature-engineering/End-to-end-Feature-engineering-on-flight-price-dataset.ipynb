```python
import pandas as pd
```

```python
df=pd.read_excel('flight_price.xlsx')
df.head(3)
```

"Time-based feature engineering" on 'Date' column

```python
# Extracting date
df['Date']=df['Date_of_Journey'].str.split('/').str[0].astype(int)
df.head(2)
```

```python
# Extracting month
df['Month']=df['Date_of_Journey'].str.split('/').str[1].astype(int)
df.head(2)
```

```python
df.info()
```

```python
# Extracting year
df['Year']=df['Date_of_Journey'].str.split('/').str[2].astype(int)
df.head(2)
```

```python
# Dropping 'Date_of_Journey' column as it is no longer needed
df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df.head(2)
```

"Time based feature-engineering" on 'Dep_Time' column

```python
# Extracting hour
df['Dep_hour']=df['Dep_Time'].str.split(':').str[0].astype(int)
```

```python
df.head(2)
```

```python
# Extracting minutes
df['Dep_min']=df['Dep_Time'].str.split(':').str[1].astype(int)
```

```python
df.head(2)
```

```python
# Dropping Dep_Time column as it is no longer needed
df.drop('Dep_Time', axis=1, inplace=True)
```

```python
df.head(3)
```

We will perform time-based feature engineering on the Arrival_Time column because some entries include a date while others do not. Since the arrival time is more important, we will drop the date portion, as the Date_of_Journey column has already been created separately.

```python
# Extracting Arrival hour
df['Arrival_Hour']=df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0].astype(int)
```

```python
df.head()
```

```python
# Extracting Arrival minute
df['Arrival_Min']=df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[1].astype(int)
```

```python
df.head()
```

```python
# Dropping Arrival_Time column as it is no longer needed
df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
df.head(2)
```

“Time-based feature engineering” on the Duration column.

```python
# Extracting duration hour
df['Duration_Hours']=df['Duration'].str.extract(r'(\d+)(?=h)', expand=False)
```

```python
df.head(3)
```

```python
df.tail(3)
```

```python
# Displaying dataset info 
df.info()
```

```python
df['Duration_Hours'].unique()
```

```python
# Filling NAN value
df['Duration_Hours'].fillna(0, inplace=True)
```

```python
df['Duration_Hours'].unique()
# NAN value successfully filled with zero
```

```python
# Transforming Duraton_Hours to integer type
df['Duration_Hours']=df['Duration_Hours'].astype(int)
```

```python
# Displaying dataset info for reference
df.info()
```

Duration_Hours type issue:

Duration was stored as strings like "2h 50m", so direct astype(int) failed.
Extracted only the numeric hours with str.extract(r"(\d+)(?=h)"), when we use str.extract the values which contain only minutes no 'h' where converted to NAN, so filled missing hours with 0, and then reassigned the converted values using df["Duration_Hours"] = df["Duration_Hours"].astype(int), which changed the dtype from object to int64.

```python
# Extracting duration minutes
df['Duration_Min']=df['Duration'].str.extract(r'(\d+)(?=m)', expand=False) #expand=False will return a series(a column)
# if true it will return a dataframe 
```

```python
df.head(2)
```

```python
df['Duration_Min'].unique()
```

Here also extarction creates a NAN value, we will fill this with 0, as there were no minutes

```python
# Filling NAN value with zero
df['Duration_Min'].fillna(0, inplace=True)
```

```python
# Transforming Duration_min column into an integer type
df['Duration_Min']=df['Duration_Min'].astype(int) 
```

```python
# Displaying data set info for reference 
df.info()
```

Successfully converted into integer type

```python
df.head(2)
```

```python
# Dropping the duration column as we no longer need that

df.drop('Duration',axis=1,inplace=True) 
```

```python
df.head(2)
```

Ordinal-encoding on Total_steps column

```python
df['Total_Stops'].unique()
```

```python
df.isnull().sum()
```

```python
df[df['Total_Stops'].isnull()]
```

As we can see, Total_Stops contains only one NaN value. Since the route distance (Delhi to Cochin) is not very long compared to the flight duration, we will safely treat this missing value as '1 stop'.

```python
df['Total_Stops'].unique()
```

```python
# performing ordinal coding on Total_steps column
import numpy as np
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'2 stops':2,'1 stop':1,'3 stops':3,'4 stops':4,np.nan:1})
```

```python
df.head()
```

```python
df['Total_Stops'].isnull().sum()
```

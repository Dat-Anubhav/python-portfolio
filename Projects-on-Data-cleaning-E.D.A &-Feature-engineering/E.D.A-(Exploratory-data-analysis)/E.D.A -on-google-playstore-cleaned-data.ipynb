## 3.2 Feature Information
1. App :- Name of the App
2. Category :- Category under which the App falls.
3. Rating :- Application's rating on playstore
4. Reviews :- Number of reviews of the App.
5. Size :- Size of the App.
6. Install :- Number of Installs of the App
7. Type :- If the App is free/paid
8. Price :- Price of the app (0 if it is Free)
9. Content Rating :- Appropiate Target Audience of the App.
10. Genres:- Genre under which the App falls.
11. Last Updated :- Date when the App was last updated
12. Current Ver :- Current Version of the Application
13. Android Ver :- Minimum Android Version required to run the App
14. Date :- The date on which the app was last updated
15. Month :- The month in which the app was last updated
16. Year :- The year in which the app was last updated

```python
# seperating the numerical and categorical features

numerical=df.select_dtypes(include='number').columns.to_list()
categorical=df.select_dtypes(exclude='number').columns.tolist()

# Printing numerical and categorical columns
print("Dataset has {} numerical columns : {}".format(len(numerical),numerical))
print("Dataset has {} categorical columns : {}".format(len(categorical),categorical))

# The first {} becomes len(numerical) (like 11).
# The second {} becomes numerical (the list of column names)
```

```python
df.head()
```

```python
# Proportion of count data on categorical columns
for col in categorical:
    print(df[col].value_counts(normalize=True)*100) # *100 gives percentage
    print("----------------------------------------------")

# If you do not use normalize=True, 
# then value_counts() gives only raw counts, not proportions or percentages
```

```python
## Proportion of count data on numerical columns
plt.figure(figsize=(15, 15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=1, y=1.)
fontsize=20

'''Sets the size of the figure title text to 20 points.

fontweight='bold'
Makes the figure title text bold.

alpha=0.8
Controls transparency of the title; 1 is fully solid, 0 is invisible, 0.8 is slightly transparent.

y=1.0
Sets the vertical position of the figure title; 0 is bottom, 1 is top edge of the figure, so 1.0 places it at the top.'''

# U can skip them then it work on the default setting , 
# well these fontzize=20, fonwight='bold' alpha=0.8 etc are the same values at default 

for i in range(0, len(numerical)):
    plt.subplot(5,3, i+1)
    sns.kdeplot(df[numerical[i]], shade=True, color='g')
    plt.xlabel(numerical[i])
    plt.tight_layout()
```
